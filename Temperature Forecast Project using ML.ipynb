```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model

# Modify the file path to the absolute path of your CSV file
file_path = "C:/Users/Zaina/Downloads/temperature.csv"

df = pd.read_csv(file_path)  # Reading data from the saved location in the system
df = df[['Present_Tmax', 'Present_Tmin', 'Next_Tmax', 'Next_Tmin']]
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
x = df[['Present_Tmax', 'Present_Tmin']]
y = df[['Next_Tmax', 'Next_Tmin']]

regr = linear_model.LinearRegression()
regr.fit(x.values, y)
sns.lmplot(data=df, x='Present_Tmax', y='Next_Tmax', ci=None)
sns.lmplot(data=df, x='Present_Tmin', y='Next_Tmin', ci=None)
print('intercept :', regr.intercept_)
print('coefficient :', regr.coef_)

# Performing the predictions here
predicted = regr.predict([[31.4, 21.9]])
print('Predicted [31.4,21.9] :', predicted)

predicted = regr.predict([[50, 44.7]])
print('Predicted [50,44.7] :', predicted)

predicted = regr.predict([[34.3, 27.4]])
print('Predicted [34.3,27.4] :', predicted)

predicted = regr.predict([[3, -4]])
print('Predicted [3,-4] :', predicted)
```
